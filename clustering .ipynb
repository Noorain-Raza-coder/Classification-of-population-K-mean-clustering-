```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(df2)
pred = kmeans.predict(df2)
```

```python
df2['cluster'] = pred
df2['cluster']
```

```python
def seg(str_x , str_y ,clusters):
    x=[]
    y=[]
    for i in range(clusters):
        x.append(df2[str_x][df2['cluster']==i])
        y.append(df2[str_y][df2['cluster']==i])        
        
    return x ,y
```

```python
def plot_clusters(str_x , str_y , clusters):
    plt.figure(figsize = (5,5),dpi = 120)
    x,y = seg(str_x , str_y , clusters)
    
    for i in range(clusters):
        plt.scatter(x[i],y[i],label = 'cluster{}'.format(i))
        
    plt.xlabel(str_x)
    plt.ylabel(str_y)
    plt.title(str(str_x +"Vs"+ str_y))
    plt.legend()
```

```python
plot_clusters('Indians','Foreigners',3)
```
